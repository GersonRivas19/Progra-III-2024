Desarrolle la tabla de meses acumulados a n meses:

```python
def acumulado_meses(n):
    return (n * (n + 1)) // 2

n = int(input("Ingrese el número de meses: "))
resultado = acumulado_meses(n)
print(f"La suma acumulada de los primeros {n} meses es: {resultado}")
```
